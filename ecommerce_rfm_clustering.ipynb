# Metodologia

A metodologia adotada para guiar a condução deste projeto de análise de dados foi o CRISP-DM (Cross-Industry Standard Process for Data Mining). O CRISP-DM é um modelo de processo amplamente reconhecido que orienta de maneira sistemática as etapas necessárias para realizar com sucesso projetos de mineração de dados e análise de dados.

O processo CRISP-DM é composto por seis etapas interconectadas, cada uma com um propósito específico:

1. **Compreensão do Negócio:** Nesta fase inicial, buscamos uma compreensão profunda dos objetivos do projeto e das necessidades do negócio. Definimos claramente o que se pretende alcançar com a análise de dados e como isso se relaciona com os objetivos maiores da organização.

2. **Compreensão dos Dados:** Durante essa etapa, coletamos e exploramos os dados disponíveis. Avaliamos a qualidade dos dados, identificamos suas fontes e entendemos as características fundamentais. Isso nos ajuda a determinar se os dados são adequados para o propósito do projeto.

3. **Preparação dos Dados:** Aqui, os dados são processados, limpos e transformados. Lidamos com desafios como valores ausentes, inconsistências e redundâncias. Também selecionamos os atributos mais relevantes e os preparamos para a análise.

4. **Modelagem:** Nesta fase, selecionamos as técnicas de modelagem apropriadas com base nos objetivos do projeto. Construímos e ajustamos modelos aos dados preparados. Isso pode envolver algoritmos de aprendizado de máquina, estatísticas ou outras abordagens analíticas.

5. **Avaliação:** Os modelos construídos são avaliados para determinar sua eficácia e relevância. Testamos a capacidade dos modelos de generalizar para novos dados e medimos sua performance em relação aos objetivos do projeto.

6. **Implantação:** Uma vez que os modelos são validados, eles são implementados em um ambiente de produção. Isso inclui a integração dos modelos com sistemas existentes, a configuração de monitoramento contínuo e a preparação para o uso prático.

##RFV - Recência, Frequência e Valor Monetário

A análise RFV é uma técnica fundamental na área de análise de clientes e marketing. Ela permite avaliar o comportamento e o valor dos clientes com base em três dimensões cruciais: Recência (R), Frequência (F) e Valor Monetário (V). Essa abordagem oferece insights valiosos para segmentar, compreender e atender de maneira eficaz diferentes grupos de clientes.

**Recência (R):**
A dimensão Recência refere-se à "Recency" em inglês, ou seja, o quão recente foi a última interação do cliente com a empresa. Isso pode incluir a última compra, a última visita ao site ou a última interação em geral.

**Frequência (F):**
A dimensão Frequência mede quantas vezes um cliente interagiu com a empresa durante um determinado período. Pode ser a frequência de compras, visitas ao site, participação em eventos, entre outros.

**Valor Monetário (V):**
A dimensão Valor Monetário avalia o quanto um cliente gasta ou contribui financeiramente para a empresa. Isso inclui o valor total das compras realizadas, assinaturas ou qualquer outra forma de transação financeira.

```python
#Criando a variável com o preço total, multiplicando a quantidade pelo preço unitário.

df['price_total'] = df.Quantity * df.UnitPrice
df.head(3)
```

```python
df['InvoiceDate'].max()

#Se a ultima venda aconteceu em dezembro de 2012, irei usar o primeiro dia de 2013 como data de referência.
```

```python
#Criando um df com apenas os valores de RFM e o id do cliente como index.

df_rfm = (
  df.groupby('CustomerID')
  .agg(
      R = ('InvoiceDate', lambda x: (pd.Timestamp('2012-01-01') - x.max()).days),
      F = ('InvoiceNo', 'nunique'),
      M = ('price_total', 'mean')
  )
)

df_rfm.head()
```

```python
# Normalizando o df_rfm, e analisando sua distribuição.

display(df_rfm.apply(scale).head(3))

#Visualizando o boxplot
df_rfm.apply(scale).plot.box()
plt.show()
```

```python
# Normalizando o df_rfm, pois as variáveis são de escalas diferentes.

scaler = PowerTransformer()
df_rfm_scaled = pd.DataFrame(scaler.fit_transform(df_rfm), index=df_rfm.index, columns=df_rfm.columns)
df_rfm_scaled.head(3)
```

```python
# Visualizando a distribuição do rfm após normalização.

df_rfm_scaled.plot.box()
```

```python
# criando um clip no quartil de 95%

df_rfm_clip = df_rfm.apply(lambda x: x.clip(upper=x.quantile(.95)))
display(df_rfm_clip.describe())

# Normalizando o df rfm clip
df_rfm_clip_scaled = df_rfm_clip.apply(scale)
```

# Modelagem

No nosso projeto, estamos utilizando o K-Means por conta de suas características favoráveis e sua adequação aos nossos objetivos. O K-Means é simples de entender e implementar. Ele é eficiente em termos computacionais, o que o torna ideal para lidar com grandes volumes de dados. Isso permite que nosso projeto seja executado de forma mais rápida e escalável.

```python
import plotly.express as px

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from yellowbrick.cluster import KElbowVisualizer
```

```python
# Usei o método Elbow para visualizar o score de distorção, e assim definir o numero ideal de clusters.
## Utilizarei quatro clusters, pois até ele o distortion score e o fit time caem consideravelmente.

visualizer = KElbowVisualizer(KMeans(), k=10)
visualizer.fit(df_rfm_clip_scaled)
visualizer.show()
```

```python
# Criei uma tabela para auxiliar na escolha do k, pois ela avalia algumas métricas importantes.

cluster_metrics = silhouette_score, davies_bouldin_score, calinski_harabasz_score
cluster_metrics_results = []
X = df_rfm_clip_scaled.copy()

for k in range(2,11):
  model = KMeans(n_clusters=k, random_state=0)
  labels = model.fit_predict(X)
  cluster_results_dict = {'k': k}
  cluster_results_dict['inertia'] = model.inertia_
  for metric in cluster_metrics:
    cluster_results_dict[metric.__name__] = metric(X, labels)
  cluster_metrics_results.append(cluster_results_dict)

pd.DataFrame(cluster_metrics_results).set_index('k').style.background_gradient()
```
